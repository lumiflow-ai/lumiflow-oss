```python
from pydantic import BaseModel

class Animal(BaseModel):
    scientific_name: str
    phylum: str
    order: str
    family: str
    genus: str
    species: str
    number_of_legs: int
    tail: bool
```

## Func

```python
# The Func model is a fake model that allows you to define a function that generates a response. 
# This is useful for testing and debugging your code without having to call an actual model.
model_func_1 = models.Func(
    function=lambda text: f"Yes, that is a good question, {text.lower()}",
)
with measure.time():
    response_func_1 = model_func_1.generate(
        "What is a dog?",
        # max_new_tokens=512,
        # temperature=0.5,
        # top_p=0.9,
    )
print("response_func_1:\n", response_func_1)
```

## VertexAI model

```python
# Set the API key for the Vertex AI model.
model_vertex_ai = models.VertexAI("google/gemini-2.5-flash")
```

```python
with measure.time():
    response_vertex_ai = model_vertex_ai.generate(
        "What is a dog?",
        stream=True,
        # max_new_tokens=512,
        # temperature=0.5,
        # top_p=0.9,
    )
    response_vertex_ai_structured_output = model_vertex_ai.generate(
        "What is a dog?",
        output_model=Animal,
        # max_new_tokens=512,
        # temperature=0.5,
        # top_p=0.9,
    )
print("response_vertex_ai:\n", ''.join(list(response_vertex_ai)))
print("response_vertex_ai_structured_output:\n", type(response_vertex_ai_structured_output), response_vertex_ai_structured_output)
```
